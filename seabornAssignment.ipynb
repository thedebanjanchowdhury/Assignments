#### Q1. Name any five plots that we can plot using seaborn, also state the uses of each one.

These are most used plots in seaborn:
1. Histogram: Used to visualize distribution of a single variable, used to understand distribution of ages ina a population.
2. Scatter Plot: Visualize to represent relationship between two numerical variables.
3. Bar Plot: Used to display relationship between categorical variable and a numerical variable.
4. Box Plot: Useful for visualizing the distribution or spread of data
5. Heatmap:  Used to visualize the correlation between variables in a dataset. 

#### Q2. Loading data and ploting the graph

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri_data = sns.load_dataset('fmri')
```

```python
sns.lineplot(x = 'timepoint', y = 'signal', hue = 'event', style = 'region', data = fmri_data)
```

#### Q3. Loading Titanic Dataset and ploting graph

```python
titanic_dataset = sns.load_dataset('titanic')

plt.subplot(1,2,1)
plt.title('Age & Pclass')
sns.boxplot(x = 'pclass', y = 'age', data=titanic_dataset)

plt.subplot(1,2,2)
plt.title('Fare & Pclass')
sns.boxplot(x = 'pclass', y = 'fare', data = titanic_dataset)

plt.tight_layout()
plt.show()
```

#### Q4. Ploting histogram using diamonds dataset

```python
diamond_dataset = sns.load_dataset('diamonds')

sns.histplot(x = 'price', data = diamond_dataset, hue = 'cut')
```

#### Q5. Ploting pair plot from iris dataset

```python
iris_dataset = sns.load_dataset('iris')

sns.pairplot(data=iris_dataset, hue = 'species')
```

#### Q6. Ploting data from flights dataset

```python
flights_data = sns.load_dataset('flights')

flights_pivot = flights_data.pivot_table(index='month',columns='year', values='passengers')
plt.figure(figsize=(10,6))
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap = 'coolwarm')

plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Passenger Count Heatmap (Flights)")
plt.show()
```
