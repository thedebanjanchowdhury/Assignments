### Q1. What is Matplotlib? why it is used? Name five plots that can be ploted using Matplotlib

Matplotlib is a python library, that is used to visualize data by drawing different charts according to the given data.

The mais usage of matplotlib is to visualize the data to get a better understanding about what the dataset contains.

Five plots which can be plotted using matplotlib are bar chart, scatter chart, pie chart, histograms and line plots 

### Q2.What is scatter plot? Use the code to generate a scatter plot

Scatter plot is a type of chart that is used to visualize data in matplotlib. The independent variable is plotted in X axis and dependent variable is plotted in Y axis.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0,2,50)
y = 3 + np.random.normal(0,2,len(x))

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Title Here')
plt.scatter(x,y)
plt.show()
```

### Q3. Why is subplot() function is used? Draw four line plots using the subplot() function.

Subplot() function is used to basically plot multiple plots inside the same graph space.

```python
x = np.array([0,1,2,3,4,5])
y1 = np.array([0,100,200,300,400,500])
y2 = np.array([50,20,40,20,60,70])
y3 = np.array([10,20,30,40,50,60])
y4 = np.array([200,350,250,550,450,150])

plt.subplot(3,3,1)
plt.plot(x,y1, color = 'red')
plt.title("First Line")

plt.subplot(3,3,2)
plt.plot(x,y2, color = 'blue')
plt.title('Second Line')

plt.subplot(3,3,3)
plt.plot(x,y3, color = 'green')
plt.title('Third Line')

plt.subplot(3,3,4)
plt.plot(x,y4, color = 'orange')
plt.title('Fourth Line')

plt.tight_layout()
plt.show()
```

### Q4. What is bar plot? Why it is used? Use the following code to generate bar plot.

- Bar plot is a visualization technique that is used to represent categorical data. In it, given data is represented in bar forms.
- Bar plot is commonly used for visualizing survey results, showing frequency distributions etc.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.title('Profits Generated by Tech Giants')
plt.xlabel("Company Name")
plt.ylabel("Profit Margin")
plt.bar(company,profit)
plt.show()
```

### Q5. What is box plot? Why is it used?

- Box plot is used to visualize data from a given dataset, to get better understanding and study trends from the data
- Box plot is mainly used to visualize spread, central tendencies from the data

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90,20,200)

data = [box1, box2]
labels=['Box1', 'Box2']

plt.boxplot(data, labels = labels) 
plt.title('boxplot example')
plt.ylabel('Values')
plt.show()
```
