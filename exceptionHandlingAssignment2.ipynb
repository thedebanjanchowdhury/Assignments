Q3. The basic two error defined in ArithmeticError class are **OverflowError** and **ZeroDivisionError**

OverflowError raises when the outcome value is to large to be expressed between the numeric expressions of the datatype

ZeroDivisionError raises when division/modulo operation is called while denominator value is zero.

```python
try:
    result1 = 10/0 #Display ZeroDivisionError
    
    print(result1)
    
except ArithmeticError as e:
    print(e)
except ZeroDivisionError as e:
    print(e)
```

Q4. Lookup operation involves searching for a value in collection of a list, dictionary of set. The primary purpose of lookuperror is to categorize various lookup related error inside one subclass

- **KeyError** : The keyerror is raised when you try to access a key that doesnot belong to a dictionary.

```python
try:
    my_dict = {'name' : 'Debanjan', 'Age' : 30}
    print(my_dict['city'])
    
except KeyError as e:
    print("Exception Arised: ", e)
```

- **IndexError** : IndexError is raised when we try to access some specific index elements from an array/list, which are not been created at all.

```python
my_list = [1, 2, 3, 4, 5]

try:
    print(my_list[10])
    
except IndexError as e:
    print("IndexError: ", e)
```
